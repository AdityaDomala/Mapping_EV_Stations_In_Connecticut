# Mapping_EV_Stations_In_Connecticut


```python
import pandas as pd
```

```python
df = pd.read_csv('Electric_Vehicle_Charging_Stations.csv')
```

```python
df
```

```python
df.dtypes
```

```python
df2 = df.drop(['Street Address','City','Access Days Time','EV Level1 EVSE Num','EV Level2 EVSE Num','EV DC Fast Count','EV Other Info'],axis = 1)
```

```python
df2.head()
```

```python
df2['New Georeferenced Column'] = df2['New Georeferenced Column'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
df2.head()
```

```python
df2[['longitude', 'latitude']] = df2['New Georeferenced Column'].str.split(' ', expand=True)
```

```python
df2.head()
```

```python
df2 = df2.drop(['New Georeferenced Column'], axis = 1)
```

```python
df2 = df2.rename(columns = {'Station Name':'station_name'})
```

```python
df2.head(2)
```

```python
df2.dtypes
```

```python
df2['longitude'] = df2['longitude'].astype(float)
```

```python
df2['latitude'] = df2['latitude'].astype(float)
```

```python
df2.dtypes
```

```python
import folium as fl
```

```python
map = fl.Map(location=['41.62133140914617', '-72.74908078449317'], zoom_start = 9)
```

```python
map
```

```python
fl.Marker([41.072882,-73.476469], popup = 'BMW OF DARIEN').add_to(map)

map
```

```python
df2.head()
```

```python
map = fl.Map(location=[df2['latitude'].mean(), df2['longitude'].mean()], zoom_start=10)

for idx, row in df2.iterrows():
    fl.Marker([row['latitude'], row['longitude']], popup=row['station_name']).add_to(map)
map
```
